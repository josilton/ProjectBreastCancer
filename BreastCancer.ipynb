Projato para analise de imagens de mamografia.

Desenvolvimento de um algoritimo para ler a imagem, realizar analise da estrutura de dados, rastrear e localizar regiões de interesse nas imagens e classificar esses achados.

    Imagem.
    No processo de análise das imagens, um dos primeiros passos é a avaliação onde perora-se identificar a distribuição dos níveis de cinza da imagem. A distribuição dos pixels com níveis de cinza de uma imagem pode apresentar sobre o nível de contraste que a imagem contém.
    
    As imagens, no processo de aquisição, são geradas com as 3 bandas de cores do espectro da luz visível (Rad, Green e Blue) conhecido como escala RGB. 

    O código abaixo carregará uma imagem, em seguida será gerará uma nova imagem em escala de cinza. 
    Em seguida será gerado um histograma das imagens na escala de cinza.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
#import matplotlib.image as mpimg
from skimage.color import rgb2gray
from skimage import exposure 
from skimage import io

#Informar o caminho da pasta que estão as fotos.
PatImg = "./BreastCancerImage/"
arq = os.listdir(PatImg)

for i in arq:
  a = PatImg + '/' + i
  #ler o arquivo
  img = io.imread(a)
  grayscale = rgb2gray(img)
  img_ar = np.array(grayscale)

    
  # mostrar a imagem
  fig, axes = plt.subplots(1,2,figsize=(8, 4),sharex=True, sharey=True)
  ax = axes.ravel()
  ax[0].imshow(img) # plota a imagem original
  ax[0].set_title("Imagem Original")
  ax[1].imshow(grayscale, cmap=plt.cm.gray) # plota a imagem em ton de cinza
  ax[1].set_title("Imagem Grayscale")
  fig.tight_layout()
  plt.show()

#Fim
```

    Resultado.
    
    As análises das imagens para o processo de aprendizado de máquina será usado na escala de cinza como podemos observar na imagem acima.
